```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize']=(20,20)
```

```python
img=cv2.imread('IMG_20191127_162607_725.jpg')
```

```python
def display(img,cmap=None):
    fig = plt.figure(figsize=(20,20))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap)
```

```python
lower_res=cv2.pyrDown(img)
```

```python
upper_res=cv2.pyrUp(img)
```

```python
display(img)
```

```python
display(lower_res)
```

```python
display(upper_res)
```

```python
img_infer=cv2.resize(img,None,fx=3,fy=3,interpolation=cv2.INTER_NEAREST)
img_infer1=cv2.resize(img,None,fx=3,fy=3,interpolation=cv2.INTER_LINEAR)
img_infer2=cv2.resize(img,None,fx=3,fy=3,interpolation=cv2.INTER_CUBIC)
img_infer3=cv2.resize(img,None,fx=3,fy=3,interpolation=cv2.INTER_LANCZOS4)
```

```python
plt.subplot(2,2,1),plt.imshow(img_infer)
plt.subplot(2,2,2),plt.imshow(img_infer1)
plt.subplot(2,2,3),plt.imshow(img_infer2)
plt.subplot(2,2,4),plt.imshow(img_infer3)
```

```python
i=cv2.imread('IMG_20191127_162607_725.jpg')
img_gray = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
display(img_gray)
```

```python
blurred_img = cv2.GaussianBlur(img_gray,(5,5),10)
display(blurred_img)
```

```python
canny=cv2.Canny(blurred_img,100,200)
display(canny)
```

```python
lap=cv2.Laplacian(blurred_img,cv2.CV_64F)
edge=cv2.convertScaleAbs(lap)
display(edge)
```

```python
img_sohx=cv2.Sobel(blurred_img,cv2.CV_8U,1,0,ksize=3)
img_sohy=cv2.Sobel(blurred_img,cv2.CV_8U,0,1,ksize=3)
img_soh=img_sohx+img_sohy
display(img_soh)
```

```python
plt.subplot(2,2,1),plt.imshow(canny)
plt.subplot(2,2,2),plt.imshow(img_sohx)
plt.subplot(2,2,3),plt.imshow(img_sohy)
plt.subplot(2,2,4),plt.imshow(img_soh)
```
