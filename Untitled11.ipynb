```python
History = cv2.imread(r'C:\Users\Mahad Shbeeb\Downloads\ImP (1)\ImP\History.jpg')          
```

```python
History_rgb = cv2.cvtColor(History, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(History_rgb)
```

```python
z=History_rgb.reshape((-1,3))
```

```python
z=np.float32(z)
```

```python
criteria=(cv2.TERM_CRITERIA_EPS+cv2.TERM_CRITERIA_MAX_ITER,30,1.0)
k=8
ret,label,center=cv2.kmeans(z,k,None,criteria,10,cv2.KMEANS_RANDOM_CENTERS)
```

```python
center = np.uint8(center)

res = center[label.flatten()]
res2=res.reshape(History_rgb.shape)
```

```python
figure, ax = plt.subplots(1, 2, figsize=(16, 8))

ax[0].imshow(History_rgb,cmap='gray')
ax[1].imshow(res2,cmap='gray')
```
