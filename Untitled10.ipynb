```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize']=(20,20)
```

```python
import cv2
whole = cv2.imread('2.jpeg')          
part = cv2.imread('1.jpeg')
```

```python
whole = cv2.cvtColor(whole, cv2.COLOR_BGR2GRAY)
part = cv2.cvtColor(part, cv2.COLOR_BGR2GRAY)
```

```python
figure, ax = plt.subplots(1, 2, figsize=(16, 8))

ax[0].imshow(whole,cmap='gray')
ax[1].imshow(part,cmap='gray')
```

```python
sift = cv2.xfeatures2d.SIFT_create()
keypoints_1, descriptors_1 = sift.detectAndCompute(whole,None)
keypoints_2, descriptors_2 = sift.detectAndCompute(part,None)
```

```python
bf = cv2.BFMatcher(cv2.NORM_L1, crossCheck=True)
```

```python
matches = bf.match(descriptors_1,descriptors_2)
matches = sorted(matches, key = lambda x:x.distance)
```

```python
img = cv2.drawMatches(part,keypoints_1,whole,keypoints_2,matches[:50],whole, flags=2)
plt.imshow(img),plt.show()
```

```python
i=cv2.imread('sunflower.png',0)
hog = cv2.HOGDescriptor()
winStride = (8, 8)
padding = (8, 8)
locations = ((10,20),(30,30),(50,50),(70,70),(90,90),(110,110),(130,130),(150,150),(170,170),(190,190))
hist = hog.compute(i, winStride, padding, locations)
```

```python
figure, ax = plt.subplots(1, 2, figsize=(16, 8))

ax[0].imshow(i,cmap='gray')
ax[1].plot(hist)
```
