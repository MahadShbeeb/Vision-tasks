```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline
plt.rcParams['figure.figsize']=(20,20)
```

```python
img=cv2.imread('s.jpg')
img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
img_gray=cv2.cvtColor(img_rgb,cv2.COLOR_RGB2GRAY)
plt.imshow(img_gray)
```

```python
edges=cv2.Canny(img_gray,20,150)
```

```python
lines=cv2.HoughLines(edges,1,np.pi/180,200)
```

```python
for i in range(len(lines)):
    rho=lines[i][0][0]
    theta=lines[i][0][1]
    a=math.cos(theta)
    b=math.sin(theta)
```

```python
x0=a*rho
y0=b*rho
```

```python
x1=int(x0+1000*(-b))
y1=int(y0+1000*(a))
x2=int(x0-1000*(-b))
y2=int(y0-1000*(a))
```

```python
cv2.line(img,(x1,y1),(x2,y2))
```

```python
cv_img=cv2.imread('IMG_20191127_162619_955.jpg')
img_rgb=cv2.cvtColor(cv_img,cv2.COLOR_BGR2RGB)
```

```python
gray=cv2.cvtColor(img_rgb,cv2.COLOR_RGB2GRAY)
plt.imshow(gray)
```

```python
circles=cv2.HoughCircles(img_gray,cv2.HOUGH_GRADIENT,1,20,param1=100,param2=25,minRadius=0,maxRadius=0)
```

```python
circles=np.uint16(np.round(circles))
```

```python
for i in circles[0,:]:
    cv2.circle(cv_img,(i[0],i[1]),i[2],(0,255,0),2)
    cv2.circle(cv_img,(i[0],i[1]),i[2],(0,0,255),3)

plt.imshow(cv_img)
```

```python
img=cv2.imread('chess.png')
img=cv2.resize(img,(600,600))
img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img_gray=cv2.cvtColor(img_rgb,cv2.COLOR_RGB2GRAY)
img_gray=np.float32(img_gray)
# plt.imshow(img_rgb)
```

```python
dst=cv2.cornerHarris(float(img_gray),9.5,0.05)
dst=cv2.dilate(dst,None)
plt.imshow(dst)
img[dst>0.01*dst.max()]=(0,0,255)
plt.imshow(img)
```

```python
img=cv2.imread('IMG_20191127_162611_045.jpg')
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
sift=cv2.xfeatures2d.SIFT_create(200)
kp=sift.detect(gray,None)
img=cv2.drawKeypoints(gray,kp,img,None,4)
plt.imshow(img)
```
