```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize']=(40,40)
```

```python
img1 = cv2.imread('part.jpg',0)          
img2 = cv2.imread('whole.jpg',0)
```

```python
plt.imshow(img1,cmap='gray')
```

```python
plt.imshow(img2,cmap='gray')
```

```python
orb = cv2.ORB_create() 

kp1, des1 = orb.detectAndCompute(img1,None)
kp2, des2 = orb.detectAndCompute(img2,None)
```

```python
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
```

```python
matches = bf.match(des1,des2)
matches = sorted(matches, key = lambda x:x.distance)
```

```python
img3 = cv2.drawMatches(img1,kp1,img2,kp2,matches[:10],None, flags=2)
plt.imshow(img3)
plt.show()
```

```python
sift = cv2.xfeatures2d.SIFT_create()

kp1, des1 = sift.detectAndCompute(img1,None)
kp2, des2 = sift.detectAndCompute(img2,None)
```

```python
# BFMatcher with default params
bf = cv2.BFMatcher()
matches = bf.knnMatch(des1,des2, k=2)
```

```python
# Apply ratio test
good = []
for m,n in matches:
    if m.distance < 0.75*n.distance:
        good.append([m])

# cv2.drawMatchesKnn expects list of lists as matches.
img3 = cv2.drawMatchesKnn(img1,kp1,img2,kp2,good,None,flags=2)

plt.imshow(img3),plt.show()
```

```python
def compute_hog(image):
    hog = cv2.HOGDescriptor()
    winStride = (8, 8)
    padding = (8, 8)
    locations = ((10,20),(30,30),(50,50),(70,70),(90,90),(110,110),(130,130),(150,150),(170,170),(190,190))
    hist = hog.compute(image, winStride, padding, locations)
    return hist 
```

```python
hist_img=cv2.imread('sunflower.png',0)
```

```python
plt.imshow(hist_img,cmap='gray')
```

```python
hist=compute_hog(hist_img)
plt.plot(hist)
```
